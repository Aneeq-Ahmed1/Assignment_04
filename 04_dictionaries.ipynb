```python
print("00_count_nums")

def count_numbers():
  count_dict = {}

  while True:
    num = input("Enter a number (or 'Exit' to quit): ")
    if num.lower() == 'Exit':
      break
    if num.isdigit():
      num = int(num)
      count_dict[num] = count_dict.get(num, 0) + 1
      print(count_dict)
    else:
      print("Invalid input. Please enter a number or 'Exit'.")
      return count_dict

def display_counts(count_dict):
  print("\n Number Counts:")
  for key,value in count_dict.items():
      print(f'{key} appears {value} times')


if __name__ == "__main__":
  count = count_numbers()
  display_counts(count)
```

```python
print("01_phonebook")

def add_contact(phonebook):
    """Function to add a contact to the phonebook."""
    name = input("Enter contact name: ").strip()
    number = input("Enter contact number: ").strip()

    if name in phonebook:
        print(f"‚ùå {name} already exists in the phonebook.")
    else:
        phonebook[name] = number
        print(f"‚úÖ {name} added to the phonebook.")

def search_contact(phonebook):
    """Function to search a contact in the phonebook."""
    name = input("Enter contact name to search: ").strip()

    if name in phonebook:
        print(f"üìû {name}: {phonebook[name]}")
    else:
        print(f"‚ùå {name} not found in the phonebook.")

def delete_contact(phonebook):
    """Function to delete a contact from the phonebook."""
    name = input("Enter contact name to delete: ").strip()
    if name in phonebook:
        del phonebook[name]
        print(f"‚úÖ {name} deleted from the phonebook.")
    else:
        print(f"‚ùå {name} not found in the phonebook.")

def display_contact(phonebook):
    """Function to display all contacts in the phonebook."""
    if phonebook:
        print("\nüì± Phonebook Contacts list:")
        for name, number in phonebook.items():
            print(f"{name.capitalize()}: {number}")
    else:
        print("‚ùå Phonebook is empty.")

def main():
    """Main function to interact with the user and manage the phonebook."""
    phonebook = {}

    while True:
        print("\nüìö Phonebook Menu")
        print("1. Add Contact")
        print("2. Search Contact")
        print("3. Delete Contact")
        print("4. Display Contacts")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ").strip()

        if choice == '1':
            add_contact(phonebook)
        elif choice == '2':
            search_contact(phonebook)
        elif choice == '3':
            delete_contact(phonebook)
        elif choice == '4':
            display_contact(phonebook)
        elif choice == '5':
            print("üëã Exiting Phonebook. Goodbye!")
            break
        else:
            print("‚ùå Invalid choice. Please enter a number between 1 to 5.")

if __name__ == "__main__":
    main()
```

```python
print("02_pop_up_shop")

def calculate_total_cost():
    fruits_price = {
        "apple": 5.0,
        "durian": 15.0,
        "kiwi": 8.0,
        "jackfruir": 12.0,
        "mango2": 6.0,
        "orange": 10.0
    }

    total_cost = 0

    for fruit, price in fruits_price.items():
        while True:
            try:
                quantity = int(input(f'How many {fruit} do you want?: '))
                if quantity < 0:
                    print("Invalid input. Please enter a non-negative number.")
                    continue
                total_cost += price * quantity
                break
            except ValueError:
                print("Invalid input. Please enter a valid number.")

    print(f"\nYour total cost is: ${total_cost:.2f}")

if __name__ == "__main__":
    calculate_total_cost()
```

```python
print("03_powerful_passwords")

import hashlib

def hash_password(password):
  return hashlib.sha256(password.encode()).hexdigest()

stored_logins = {
    "ali@yahoo.com":hash_password("123456"),
    "admin@yahoo.com":hash_password("654321")
}

def login(email,password):
  if email in stored_logins:
    return stored_logins[email] == hash_password(password)
  return False


if __name__ == "__main__":
  email = input("Enter your email: ")
  password = input("Enter your password: ")


  if login(email,password):
    print("Login successful!")
  else:
    print("Invalid email or password.")
```
